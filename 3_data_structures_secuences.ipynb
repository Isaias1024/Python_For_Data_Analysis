```python
#Zip
'''Pairs los elementos de listas, tuplas o cualquier otra secuencia '''
seq1 = ['chivas','america','cruz azul']
seq2 = ['12','13','9']

zipped = zip(seq1,seq2)

"se puede imprimir con un for btw para probar antes"
"para comprobar podemos castearlo a una lista e iterar y veremos"
list(zipped)
    
    
"Zip tambien puede tomar puede tomar numeros arbitrarios como secuencia, and the number of elements it produces is determined by the shortest sequence:"
seq3 = [False, True]
list(zip(seq1,seq2,seq3))

"un uso muy comun de zip es iterar simultaneamente en multiples secuencias, y posiblemente combinadas con enumarate:"
for i , (a,b) in enumerate(zip(seq1,seq2)):
    print('{0}: {1},{2}'.format(i,a,b))
    
"cuando con nos dan un zipped, zip es buena idea tu 'Unzip' la secuencia, "
futbolistas = [('Messi','Argentina'),('Cr7','Portugal')]
nombre, equipo = zip(*futbolistas)

print('Jugadores:{0} Paises:{1}'.format(nombre,equipo))
```

```python
#Lambda funcion
'Una forma de escribir una funcion en una stament y el resultado es el valor resultante, definida con la keyword lambda, no tienen significado alguno'
def al_doble(x):
    return x * 2

equivalente_line = lambda x: x*2

'Son muy utilizadas y convenientes para el analisis de datos, como veremos despues, existiran muchos casos donde tomaremos funciones como argumentos'
'es usualmente menos typing y limpio pasar una funcion landas as opposed a escribir toda una full-out delcaracion de una funcion'

'pasamos una lista y una mini funcion'
'la funcion lambda se aplica a x antes de realizar la iteracion'
def apply_to_list(some_list,f):
    return [f(x) for x in some_list]

inst = [4,0,1,5,6]
apply_to_list(inst, lambda x:x*2)
```

```python
#Currying: partial argument application
"Curryin es cumputer scienve jargon, deriva una funcion sobre una existete by argumentos parciales de aplicacioon"

'ejemplo, tenemos una funcion para sumar'
def sum(a,b):
    return a+b

'podemos derive a new function to of one variable, add_five, que agrega 5 to its argument'

add_five = lambda y: sum(5,y)

'el segundo argumento sum() is llamado curried, una nueva funcion que llama a una funcion existente.'
'Existe built-in llamado functools para simplificar todos esto'

from functools import partial
add_five = partial(sum, 5)
```
